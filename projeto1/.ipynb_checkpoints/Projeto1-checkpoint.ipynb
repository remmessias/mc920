```python
import matplotlib.pyplot as plt

import numpy as np

from PIL import Image
```

```python
from PIL import Image

def sepia(image_path:str)->Image:
    img = Image.open(image_path)
    width, height = img.size

    pixels = img.load() # create the pixel map

    for py in range(height):
        for px in range(width):
            r, g, b = img.getpixel((px, py))

            tr = int(0.393 * r + 0.769 * g + 0.189 * b)
            tg = int(0.349 * r + 0.686 * g + 0.168 * b)
            tb = int(0.272 * r + 0.534 * g + 0.131 * b)

            if tr > 255:
                tr = 255

            if tg > 255:
                tg = 255

            if tb > 255:
                tb = 255

            pixels[px, py] = (tr,tg,tb)

    return img

sepia_image = sepia('../projeto1/imagens_coloridas/baboon.png')

plt.imshow(np.asarray(sepia_image))
```

```python
def rgb2gray(image_path:str):
    img = np.asarray(Image.open(image_path))
    
    return np.dot(img[...,:3], [0.1440, 0.5870,0.2989 ])

gray_image = rgb2gray('../projeto1/imagens_coloridas/baboon.png')

plt.imshow(np.asarray(gray_image), cmap = "gray")
```

```python
import cv2
from skimage import io

img = Image.open('../projeto1/imagens_cinzas/butterfly.png')
```

```python
filter_h1 = np.array([
    [-1, 0 , 1],
    [-2, 0 , 2],
    [-1, 0, 1]
]).astype(float)

image_h1 = cv2.filter2D(np.asarray(img), -1, filter_h1)
image_h1_normalized = np.copy(image_h1)
cv2.normalize(image_h1, image_h1_normalized, 0, 255, cv2.NORM_MINMAX)
io.imshow(image_h1_normalized.astype(int), cmap='gray', vmin=0, vmax=255)
plt.title("Filtro H1 (Sobel Vertical)")
plt.show()
```

```python
filter_h2 = np.array([
    [-1, -2 , -1],
    [0, 0 , 0],
    [1, 2, 1]
]).astype(float)

image_h2 = cv2.filter2D(np.asarray(img), -1, filter_h2)
image_h2_normalized = np.copy(image_h2)
cv2.normalize(image_h2, image_h2_normalized, 0, 255, cv2.NORM_MINMAX)
io.imshow(image_h2_normalized.astype(int), cmap='gray')
plt.title("Filtro H2 (Sobel Horizontal)")
plt.show()
```

```python
filter_h3 = np.array([
    [-1, -1, -1],
    [-1, 8, -1],
    [-1, -1, -1]
]).astype(float)

image_h3 = cv2.filter2D(np.asarray(img), -1, filter_h3)
image_h3_normalized = np.copy(image_h3)
cv2.normalize(image_h3, image_h3_normalized, 0, 255, cv2.NORM_MINMAX)
io.imshow(image_h3_normalized.astype(int), cmap='gray')
plt.title("Filtro H3 (Detecção de bordas de objetos)")
plt.show()
```

```python
filter_h4 = np.array([
    [1, 1, 1],
    [1, 1, 1],
    [1, 1, 1]
]).astype(float) / 9

image_h4 = cv2.filter2D(np.asarray(img), -1, filter_h4)
image_h4_normalized = np.copy(image_h4)
cv2.normalize(image_h4, image_h4_normalized, 0, 255, cv2.NORM_MINMAX)
io.imshow(image_h4_normalized.astype(int), cmap='gray')
plt.title("Filtro H4 (Filtro linear de suavização)")
plt.show()
```
